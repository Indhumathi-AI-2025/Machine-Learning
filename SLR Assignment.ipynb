```python
# IMPORTING PANDAS LIBRARY
#pandas is a python package, it contains functions realted to db 
#which is used for data analysis, and
#we are importing pandas because we are going to work on dataset
#pandas is a little big word, so mentioning as pd 
#pd is easy to use in the program

import pandas as pd
```

```python
#LOADING THE CSV FILE INTO THE PANDAS LIBRARY
#We are reading the csv file by using pd and assigning it to a variable 
#called dataset

dataset = pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
#ASSIGNING THE INPUT
#We already got the preprocessed data, then the next step is spliting inputs and output
#so assigning the input in the variable named, independent
 
independent = dataset[["YearsExperience"]]
```

```python
#ASSIGNING THE OUTPUT
#assigning the output to the variable named, dependent

dependent = dataset[["Salary"]]
```

```python
#Once, Split i/p & o/p is done, we need to split train set & test set
#all the ML algorthims are provided by sklearn library, so by using a class called
#model_selection & importing the train_test_split function

from sklearn.model_selection import train_test_split
```

```python
#SPLITING TRAIN & TEST SET
#Simply calling the train_test_split() function, initially it shows an error 
#because it is a parameterized function as give the necessary parameters

train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#ASSIGNING THE TRAIN SET & TEST SET TO THE VARIABLES
#As we got the output as a list, this list returns 4 variables as output
#so we are assigning the 4 items to the 4 variables.
#Also, 1st place is allocated for train set input, 2nd place is allocated test set input
#3rd place for train set output & 4th for test set output

X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#AI MODEL CREATION
#A class called linear_model and calling LinearRegression function
#assigning the LinearRegression() to a variable called regressor
#Using the fit method, passing the train set inputs & output
#and make the AI to learn the model

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# AI MODEL IS CREATED
#Once the fit method is executed, our AI model is created
# The weight and bias value are stored in the regressor variable
```

```python
#TO KNOW THE WEIGHT VALUE
#Weight value is performed in the coef_ method
#assigning to the variable called weight

weight = regressor.coef_
```

```python
weight
```

```python
#TO KNOW THE BIAS VALUE
#Bias value is performed in the intercept_ method
#assigning to the variable called weight

bias = regressor.intercept_
```

```python
bias
```

```python
#Our AI model is created and we know the w & b value
```

```python
#The created model is given as input to the test set
#We are checking the inputs of the test set and assigning them to Y_pred

Y_pred = regressor.predict(X_test)
```

```python
#EVALUATION METRICS
#Using metrics class and importing r2_score function
#Assigning to a new variable and passing parameters
#parameters are Y_test, Y_pred (checking for test set)

from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
```

```python
#We got the R2 value
#It is nearest to 1. So our Model is performing GOOD
#We can save this model for use & check for the given input (Deployment)

r_score
```

```python
#Before checking for the given input
#We need to save the model
#Save the model using pickle library
#assign the file to a variable called filename
#save in .sav file format because pickle supports only .sav

import pickle
filename = "final_slr_model.sav"
```

```python
#We are writing the created model into the newly created file called final_slr_model

pickle.dump(regressor,open(filename,'wb'))
```

```python
#Once we write the model in the file, we need to load the model using the load function 
#and check for the user input 

loaded_model = pickle.load(open("final_slr_model.sav",'rb'))
#result = loaded_model.predict([[15]])
result = loaded_model.predict([[int(input("Enter the No of Years Experience:"))]])
```
